**Support for triangles **

A triangle is Equilateral if All three sides are equal:
a == b == c

Isosceles : Exactly two sides are equal:
(a == b or a == c or b == c) but not all three equal

Scalene : All three sides are different:
a != b and a != c and b != c




```python

# YOUR CODE HERE
# 1) Inputs a, b, c with validation
# 2) If invalid -> print("Not a triangle")
# 3) Else -> choose exactly one triangle type
try:
  print("Please provide 3 side lengths of a triangle: ")
  a = float(input("Side a: "))
  b = float(input("Side b: "))
  c = float(input("Side c: "))

  if a <= 0 or b <= 0 or c <= 0:
    raise ValueError("Negative number")
  elif a == str or b == str or c == str:
    raise ValueError
  elif a + b < c or a + c < b or b + c < a:
    print("Not a triangle.")
  elif a == b == c:
    print("Equilateral")
  elif a == b or a == c or b == c:
    print("Isosceles")
  elif a != b and a != c and b != c:
    print("Scalene")
  else:
    raise ValueError

except ValueError as e:
  if str(e) == "Negative number":
    print("Please provide a positive number.")
  else:
    print("INvalid input.")

```
